```python
def Millions_to_numeric(value):
    numeric_part=value.replace('Millions','').strip()
    return float(numeric_part)*1000000
```

```python
df['Store Sales']=df['Store Sales'].apply(Millions_to_numeric)
```

```python
df['Store Cost']=df['Store Cost'].apply(Millions_to_numeric)
```

```python
Sales_mean=df['Store Sales'].mean()
Sales_mean
```

```python
def promotion_based_on_sales(sales):
    if(sales>Sales_mean):
        return('Top-Tier Discounts')
    else:
        return ('Thrifty Deals')
```

```python
def promotion_name_replacement(row):
    if pd.isnull(row['Promotion Name']):
        return promotion_based_on_sales(row['Store Sales'])
    else:
        return row['Promotion Name']
```

```python
df['Promotion Name']=df.apply(promotion_name_replacement,axis=1)
```

```python
df['Store Kind'].isnull().sum()
```

```python
df_non_null = df.dropna(subset=['Store Kind'])
avg_sales_by_kind = df_non_null.groupby('Store Kind')['Store Sales'].mean()
avg_sales_by_kind 
```

```python
def store_kind_replacement(row):
    if pd.isna(row['Store Kind']):
        differences=avg_sales_by_kind-row['Store Sales']
        closest_kind=differences.abs().idxmin()
        return closest_kind
    return row['Store Kind']
```

```python
df['Store Kind']=df.apply(store_kind_replacement,axis=1)
```

```python
df['Store Kind'].isnull().sum()
```

```python
df['Store Kind'].value_counts()
```

```python
avg_weight_by_kind=df.groupby('Store Kind').agg({'Gross Weight':'mean','Net Weight':'mean'})
```

```python
def weight_replacement(row):
    if pd.isnull(row['Gross Weight']):
        row['Gross Weight']=avg_weight_by_kind.loc[row['Store Kind'],'Gross Weight']
    if pd.isnull(row['Net Weight']):
        row['Net Weight']=avg_weight_by_kind.loc[row['Store Kind'],'Net Weight']
    return row          
```

```python
df=df.apply(weight_replacement,axis=1)
```

```python
df[['Gross Weight','Net Weight']].isnull().sum()
```

```python
avg_area_by_kind=df.groupby('Store Kind').agg({'Store Area':'mean','Grocery Area':'mean','Frozen Area':'mean','Meat Area':'mean'})
```

```python
def area_replacement(row):
    if pd.isnull(row['Store Area']):
        row['Store Area']=avg_area_by_kind.loc[row['Store Kind'],'Store Area']
    if pd.isnull(row['Grocery Area']):
        row['Grocery Area']=avg_area_by_kind.loc[row['Store Kind'],'Grocery Area']    
    if pd.isnull(row['Frozen Area']):
        row['Frozen Area']=avg_area_by_kind.loc[row['Store Kind'],'Frozen Area']
    if pd.isnull(row['Meat Area']):
        row['Meat Area']=avg_area_by_kind.loc[row['Store Kind'],'Meat Area']
    return row    
```

```python
df=df.apply(area_replacement,axis=1)
```

```python
df['Is Recyclable?']=df['Is Recyclable?'].replace({'yes':1,'no':0})
df['Is Recyclable?']=df['Is Recyclable?'].astype('category')
```

```python
def income_to_numeric(val):
    if isinstance(val,(int,float)):
        return float(val)
    else:    
        numeric_part=val.replace('K+','').strip()
        return float(numeric_part)*1000
```

```python
df['Yearly Income']=df['Yearly Income'].apply(income_to_numeric)
```
